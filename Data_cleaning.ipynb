```python
# cleaning data -
df.isnull().sum()  # return all missing/empty spots
df.isnull()
df.dropna(axis=1)

# 1- Handling missing value !
  # Age-fill with median
age_data=df["Age"].median()
df["Age"]=df["Age"].fillna(age_data)
  # Embarked with mode 
comm_embarked=df["Embarked"].mode()[0]
df["Embarked"]=df["Embarked"].fillna(comm_embarked)


```

```python
df.isnull().sum()
```

```python
df.drop("Cabin", axis=1,inplace=True)
```

```python
df
```

```python
df.duplicated()
```

```python
# 2- Handle with duplicate values 
df.duplicated().sum()
# No duplicate value in dataset
```

```python
# 3- Handle with Datatype
df.dtypes
df["Age"]=df["Age"].astype("int64")
```

```python
df.dtypes
```

```python
df
```

```python
# 4- Handle with Strings 
df["Sex"]=df["Sex"].str.upper()
df["Name"]=df["Name"].str.upper()
```

```python
df.describe()
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")

plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
df["Survived"].value_counts().plot(kind="bar")

plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Passengers")
plt.show()
```

```python
plt.boxplot(df["Fare"])

plt.title("Fare Boxplot")
plt.show()
```
